```python
import xarray as xr
```

```python
folder = "/scratch/v46/fm6730/dataout/obs/integrated/vorticity_budget/merged/"
```

```python

variables = ["vr","div","vrD", "fD", "residual", "vr_stretch", "adv_vr_zonal", "adv_vr_meridional", "adv_vr", "adv_vr_vertical"]
# variables = ["vr_tilt"]
waves = ["er", "mjo", "kelvin"]

for wave in waves:
    for x in variables:
        print(f"Processing variable: {x}, wave: {wave}")
        ds = xr.open_dataset(folder + f"{x}.{wave}.30_truth_for_S2S.nc")
        encoding = {}
        for var in ds.data_vars:
            data = ds[var]
            # Calculate scale_factor and add_offset for optimal packing
            vmin = float(data.min())
            vmax = float(data.max())
            
            # Pack into int16 (range: -32768 to 32767)
            scale_factor = (vmax - vmin) / 65534  # Leave room for missing values
            add_offset = vmin + 32767 * scale_factor
            
            encoding[var] = {
                "zlib": True, 
                "complevel": 4,
                "dtype": "int16",
                "scale_factor": scale_factor,
                "add_offset": add_offset,
                "_FillValue": -32768,
            }
        ds.to_netcdf(folder + f"compressed_{x}.{wave}.30_truth_for_S2S.nc", encoding=encoding)
```
